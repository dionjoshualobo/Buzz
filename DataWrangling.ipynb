```python
# Check for leading/trailing spaces in column names
print("Columns with leading/trailing spaces:")
for col in data.columns:
    if col.strip() != col:
        print(col)

# Check for inconsistent spellings/capitalization in 'District' column
print("\nUnique values in 'District' column:")
print(data['District'].unique())
```

```python
data['State'].unique()
```

```python
top_5_yield_crops = data.groupby('Crop')['Yield'].mean().sort_values(ascending=False).head()
print("Top 5 highest average 'Yield' crops:")
print(top_5_yield_crops)
```

```python
# Count the occurrences of each crop
crop_counts = data['Crop'].value_counts()

# Find crops that appear fewer than 300 times
rare_crops = crop_counts[crop_counts < 300]

print("Crops appearing fewer than 300 times:")
print(rare_crops)
```

```python
# Rename columns with leading or trailing spaces
data.columns = data.columns.str.strip()

# Verify the column names have been updated
print("Updated column names:")
print(data.columns)
```

```python
# Group by 'Crop' and 'Season' and calculate the total Production and average Yield
seasonal_performance = data.groupby(['Crop', 'Season']).agg(
    Total_Production=('Production', 'sum'),
    Average_Yield=('Yield', 'mean')
)

print("Seasonal performance of each crop:")
print(seasonal_performance)
```

```python
# Pivot the data to show year-wise average yield for each crop
year_wise_yield = data.pivot_table(index='Crop', columns='Crop_Year', values='Yield', aggfunc='mean')

print("Year-wise average yield for each crop:")
display(year_wise_yield)
```

```python
# Melt the pivoted data to easily find the maximum yield
melted_yield = year_wise_yield.reset_index().melt(id_vars='Crop', value_vars=year_wise_yield.columns, var_name='Crop_Year', value_name='Yield')

# Find the row with the highest yield
highest_yield_row = melted_yield.loc[melted_yield['Yield'].idxmax()]

print("Crop and year with the highest average yield:")
print(f"Crop: {highest_yield_row['Crop']}")
print(f"Year: {int(highest_yield_row['Crop_Year'])}")
print(f"Average Yield: {highest_yield_row['Yield']}")
```

```python
display(melted_yield)
```

```python
# Convert 'State' and 'District' columns to lowercase and remove leading/trailing spaces
data['State'] = data['State'].str.lower().str.strip()
data['District'] = data['District'].str.lower().str.strip()

# Verify the changes
print("Unique values in 'State' column after cleaning:")
print(data['State'].unique())
print("\nUnique values in 'District' column after cleaning:")
print(data['District'].unique())
```

```python
# Calculate the average yield for each District and Crop combination
data['Average_District_Crop_Yield'] = data.groupby(['District', 'Crop'])['Yield'].transform('mean')
print(data['Average_District_Crop_Yield'])

imputed_prod_group = data['Area'] * data['Average_District_Crop_Yield']
print(imputed_prod_group)


data['Production'] = data['Production'].fillna(imputed_prod_group)

# Verify that the missing values in 'Production' are filled
print("Missing values in 'Production' after imputation:")
print(data['Production'].isnull().sum())
```

```python
# Filter the data for Odisha state and Wheat crop
odisha_wheat_production = data[(data['State'] == 'odisha') & (data['Crop'] == 'Wheat')]

# Calculate the total production of wheat in Odisha
total_odisha_wheat_production = odisha_wheat_production['Production'].sum()

print(f"Total Wheat Production in Odisha: {total_odisha_wheat_production}")
```

```python
# Load the external data
external_data = pd.read_csv("/content/city_day.csv")

# Display the first few rows and info of the external data to check for common columns
print("External Data Info:")
external_data.info()
print("\nExternal Data Head:")
display(external_data.head())

# Check for a common column for merging
print("\nColumns in current data:")
print(data.columns)
print("\nColumns in external data:")
print(external_data.columns)
```

```python
# Get unique values from 'City' column in both dataframes
unique_cities_data = set(data['City'].unique())
unique_cities_external = set(external_data['City'].unique())

# Combine unique values into a single set and sort them alphabetically
all_cities = sorted(list(unique_cities_data.union(unique_cities_external)))

print("List of all unique cities from both dataframes (alphabetical order):")
print(all_cities)
```

```python
# Standardize city names for better merging

# Standardize 'Hyderabad'
data['City'] = data['City'].replace('hyderabad', 'Hyderabad')
external_data['City'] = external_data['City'].replace('hyderabad', 'Hyderabad')

# Standardize 'Visakhapatnam'
data['City'] = data['City'].replace('visakhapatanam', 'Visakhapatnam')
external_data['City'] = external_data['City'].replace('visakhapatanam', 'Visakhapatnam')

# Merge the two dataframes on the 'City' column
merged_data = pd.merge(data, external_data, on='City', how='left')

# Display the first few rows of the merged data
print("Merged Data Head after standardization:")
display(merged_data.head())

# Check the shape of the merged data
print("\nMerged Data Shape after standardization:")
print(merged_data.shape)
```

```python
print("Columns in the merged data:")
print(merged_data.columns)
```

```python
# Create the 'Pollutant ratios' column
merged_data['Pollutant_ratios'] = merged_data['SO2'] / merged_data['NOx']

# Display the first few rows with the new column
print("Merged Data Head with 'Pollutant_ratios' column:")
display(merged_data.head())
```

```python
# Create a new column 'Calculated_Yield'
merged_data['Calculated_Yield'] = merged_data['Production'] / merged_data['Area']

# Calculate the sum of the difference between 'Calculated_Yield' and 'Yield'
difference_sum = (merged_data['Calculated_Yield'] - merged_data['Yield']).sum()

print(f"Sum of the difference between Calculated_Yield and Yield: {difference_sum}")
```

```python
# Check for negative values in Area, Production, or Yield
print("Checking for negative values:")
print("Negative Area:", merged_data[merged_data['Area'] < 0].shape[0])
print("Negative Production:", merged_data[merged_data['Production'] < 0].shape[0])
print("Negative Yield:", merged_data[merged_data['Yield'] < 0].shape[0])

# Check if Production > 0 and Area == 0
print("\nChecking for Production > 0 and Area == 0:")
print("Count:", merged_data[(merged_data['Production'] > 0) & (merged_data['Area'] == 0)].shape[0])

# Check if any Crop_Year is in the future (e.g., > 2024)
print("\nChecking for future Crop_Year:")
print("Future Crop_Years:", merged_data[merged_data['Crop_Year'] > 2024].shape[0])
```
